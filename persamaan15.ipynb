```python
from sympy import *
from sympy.physics.quantum import TensorProduct
gamma = Symbol("gamma",real = True)
q = Symbol("q", real = True)
r = Symbol("r", real = True)
s = Symbol("s", real = True)

A = Matrix([[gamma,sqrt(gamma*(1-gamma))],[sqrt(gamma*(1-gamma)),(1-gamma)]])
A
```

```python
B = (1/2)*Matrix([[(1-s),-(q-I*r)],[-(q+I*r),(1-s)]])
B
```

```python
M = TensorProduct(A,B)
M
```

```python
M_new = simplify(M)
M_new
```

```python
Eigen_values = M_new.eigenvals()
```

```python
keys = [k for k in Eigen_values]
```
